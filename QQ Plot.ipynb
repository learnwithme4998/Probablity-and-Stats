```python
# QQ plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
#Loc - mean Scale - Variance
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)

# 1 to 100th percentiles of std-normal
for i in range(1,101):
    print(i, np.percentile(std_normal,i))
```

```python
import matplotlib.pyplot as plt
```

```python
# generate 100 sanples from N(20,5)
testing_sample = np.random.normal(loc = 20, scale = 5, size=10000) 
#try size=1000


stats.probplot(testing_sample, dist="norm", plot=pylab)
pylab.show()
```

```python
# Using from Normal Distribution
testing_sample1 = np.random.uniform(low = -1, high = 1, size=10000) 

stats.probplot(testing_sample1, dist="norm", plot=pylab)
pylab.show()
```
